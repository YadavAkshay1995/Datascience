```python
x=Df.Waist
y=Df.AT
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(fit_intercept=True)
```

```python
model.fit(x[:,np.newaxis],y)
```

```python
xfit=x
```

```python
yfit=model.predict(xfit[:, np.newaxis])
```

```python
plt.scatter(x,y)
plt.plot(xfit,yfit)
```

```python
plt.plot(Df.Waist, Df.AT,)  # Plot the chart
plt.show() 
```

```python
#Taking x-y values from model
model.score(x[:,np.newaxis],y)
```
