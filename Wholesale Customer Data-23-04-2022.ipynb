```python
from sklearn.cluster import DBSCAN
#DBScan
dbscan=DBSCAN(eps=2,min_samples=6)
dbscan.fit(X)
```

```python
#Noisy samples are given label-1
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=["cluster"])
c1
```

```python
clustered=pd.concat([wsd,c1],axis=1)
clustered
```
